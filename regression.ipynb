```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import Regression # import Regression module you wrote (see Regression.py)
```

# Load toy dataset from scikit-learn
Contains data about Boston home prices

```python
X,y = datasets.load_boston(return_X_y=True)
```

# OLS Estimates

```python
beta_hat = Regression.OLS(X,y,demean=False)
print('Estimated coefficients: \n{}'.format(np.around(beta_hat,2)))
```

# Ridge and Lasso Regression Coefficient Estimates

```python
low = 0
high = 5
n = 20
reg_strengths = np.logspace(low, high, n)
ridge_estimates = list()
lasso_estimates = list()

for lambd in reg_strengths:
    beta_ridge = Regression.ridge(X, y, lambd)
    beta_lasso = Regression.lasso(X, y, lambd)
    ridge_estimates.append(beta_ridge[1:])
    lasso_estimates.append(beta_lasso[1:])
    
ridge_estimates = np.vstack(ridge_estimates)
lasso_estimates = np.vstack(lasso_estimates)

# Plot ridge results
plt.figure(figsize=(12,6))
plt.plot(np.linspace(low,high,n), ridge_estimates)
plt.title('Ridge Regression Coefficient Estimates', size=18)
plt.ylabel('Coefficient Values', size=14)
plt.xlabel('Log Regularization Strength -- Log({})'.format(r'$\lambda$'), size=14)
plt.grid()
plt.show()

# Plot lasso results
plt.figure(figsize=(12,6))
plt.plot(np.linspace(low,high,n), lasso_estimates)
plt.title('Lasso Regression Coefficient Estimates', size=18)
plt.ylabel('Coefficient Values', size=14)
plt.xlabel('Log Regularization Strength -- Log({})'.format(r'$\lambda$'), size=14)
plt.grid()
plt.show()
```
